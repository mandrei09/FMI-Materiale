**Exercitiul 1**

a. Citiți imaginile din aceste fișiere și salvați-le într-un np.array (va avea 
dimensiunea 9x400x600).

```python
images = []
for i in range (9):
    image_path = 'images/images/car_'+str(i)+'.npy'
    image = np.load(image_path)
    images.append(image)
images = np.array(images)

print(images)
```

b. Calculați suma valorilor pixelilor tuturor imaginilor.

```python
print('\nSuma tuturor imaginilor:',np.sum(images))
```

c. Calculați suma valorilor pixelilor pentru fiecare imagine în parte.

*Fiecare dimensiune care este adaugata tabloului **are indicele axei 0**, apoi se continua numaratoarea dimensiunilor de la tablourile inferioare ca si dimensiune.*

```python
# De exemplu:
#                          ^[[1,2,3,4]
# [0,1,2,3,4]             0|[5,6,7,8]]
# ---------->              |---------->
#      0                         1

# [
#     ^[[1,2,3] [[1,2,3]
#   1 |[4,5,6]] [4,5,6]]
# ]   --------->
#         2
#---------------------->
#           0
     
sums = np.sum(images, axis=(1,2))
print(sums)
```

d. Afișați indexul imaginii cu suma maximă.

```python
print('\nMaximum index',np.argmax(sums))
```

e. Calculați imaginea medie și afișati-o. 
Obs: Afișarea imaginii medii se poate face folosind biblioteca scikit-image în  
următorul mod: 

*La fel ca mai devreme, distrugem axa 0 pentru a calcula imaginea medie.*

```python
# from skimage import io 
# io.imshow(mean_image.astype(np.uint8)) # petru a putea fi afisata  
#                                        # imaginea trebuie sa aiba  
#                                        # tipul unsigned int 
# io.show()
```

```python
print('\nImaginea medie')
avg_img = np.mean(images, axis=0)
print(avg_img)
```

f. Cu ajutorul funcției np.std(images_array), calculați deviația standard a 
imaginilor.

*Deviatia imaginilor -> gradul de imprastiere.*

```python
print('\nDeviatia standard a imaginilor')
dev = np.std(images)
print(dev)
```

g. Normalizați imaginile. (se scade imaginea medie și se împarte rezultatul la 
deviația standard)

```python
print('\nNormalizarea imaginilor')
image2 = (avg_img)/dev
print(image2)
```

respectiv coloanele cuprinse între 280 și 400.
